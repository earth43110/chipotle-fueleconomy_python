# Python Final Project (on fuel economy data)

## Import seaborn library

```python
import seaborn as sns
```

## Load dataset

```python
mpg = sns.load_dataset("mpg")
mpg.head()
```

## Create a visual showing the relationship between horsepower and fuel economy

```python
sns.relplot(data=mpg,x='horsepower',y='mpg')
# A negtive relationship or correlation is shown here between horsepower and fuel economy
```

## Create a table summarizing fuel economy by country of origin

```python
mpg.groupby(by='origin')['mpg'].sum()
```

## Create a visual showing how fuel economy has changed over time

```python
sns.barplot(data=mpg, x='model_year', y='mpg', errorbar=None)
# The fuel economy increases as time passes
```

## Create two other visuals or tables related to fuel economy

```python
sns.relplot(data=mpg, x='mpg', y='weight')
# A negtive relationship or correlation is shown here between horsepower and fuel economy
```

```python
sns.relplot(data=mpg, x='mpg', y='displacement')
# A negtive relationship or correlation is shown here between horsepower and fuel economy
```
